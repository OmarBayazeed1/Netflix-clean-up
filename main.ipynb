```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

df=pd.read_csv('/kaggle/input/datasets/omarbayazeed/netflix/netflix_titles.csv')
```

```python
df['director']=df['director'].fillna('None')
df['country']=df['country'].fillna('None')
df['cast']=df['cast'].fillna('None')
df['date_added']=df['date_added'].fillna('None')
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
df=df.dropna(subset='duration')
```

```python
#it is important to split the mixed value and type into separate columns
df[["duration_value", "duration_type"]] = df["duration"].str.split(" ", expand=True)
```

```python
#making the value numeric not string so we can make math operaations on it 
df["duration_value"] = pd.to_numeric(df["duration_value"], errors="coerce")
```

```python
df["duration_type"] = df["duration_type"].replace({
    "min": "Minutes",
    "Season": "Seasons",
    "Seasons": "Seasons"
})
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
df.to_csv('cleaned-data.csv')
```
